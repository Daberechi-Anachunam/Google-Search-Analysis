# INTRODUCTION
Here I will be analyzing the Google search trends on the queries based on “Data Science”.

```python
import pandas as pd
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
trend = TrendReq()
```

Now I will create a DataFrame of the top 10 countries which search for “Data Science” on Google:

```python
trend.build_payload(kw_list=["Data Science"])
data = trend.interest_by_region()
data = data.sort_values(by="Data Science", ascending=False)
data = data.head(10)
print(data)
```

So, according to the above results, the search queries based on “Data Science” are mostly done in Singapore. interestinly Nigeria is in the fifth position showing that Nigerians interest in this Career path.  We can also visualize this data using a bar chart

```python
data.reset_index().plot(x="geoName", y="Data Science", 
                        figsize=(20,15), kind="bar")
plt.style.use('fivethirtyeight')
plt.show()
```

So as we all know that Data Science has been the focus of so many companies and students for the last 3-4 years, so let’s have a look at the trend of searches to see how the total search queries based on “Machine Learning” increased or decreased on Google:

```python
data = TrendReq(hl='en-US', tz=360)
data.build_payload(kw_list=['Data Science'])
data = data.interest_over_time()
fig, ax = plt.subplots(figsize=(20, 15))
data['Data Science'].plot()
plt.style.use('fivethirtyeight')
plt.title('Total Google Searches for Data Science', fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Total Count')
plt.show()
```

Conclusion
So we can see that searches based on “Data Science” on Google has been growing since 2018 and suddenlyn had an exponential growth in 2022 showing that this year that has been more people being interested in this career path. This is how we can analyze Google searches based on any keyword. A business can perform Google search analysis to understand what people are looking for on Google at any given time.
